## Arrays e operações vetorizadas

O **NumPy** oferece o tipo `ndarray`: arranjos homogêneos (todos os elementos do mesmo tipo) otimizados em memória. Em Ciência de Dados, isso permite **vetorização**: operações em lote, sem loops Python por elemento, o que costuma ser bem mais rápido e legível.

Conceitos úteis: **shape** (dimensões), **dtype** (tipo dos elementos), **eixos** em arrays multidimensionais e **broadcasting** (extensão inteligente de dimensões em operações entre arrays compatíveis).

````python
lista = [1, 2, 3, 4, 5]
a = np.array(lista, dtype=np.float64)
print("array:", a)
print("shape:", a.shape, "| ndim:", a.ndim, "| dtype:", a.dtype)

b = np.arange(12).reshape(3, 4)
print("\nmatriz 3x4:\n", b)

print("\nOperações vetorizadas (elemento a elemento):")
print("a * 10:", a * 10)
print("a ** 2:", a ** 2)
print("sqrt(a):", np.sqrt(a))

print("\nAgregações ao longo de eixos:")
print("soma total de b:", b.sum())
print("soma por coluna (eixo 0):", b.sum(axis=0))
print("média por linha (eixo 1):", b.mean(axis=1))

print("\nMáscaras booleanas (filtrar sem loop):")
print("elementos > 6:", b[b > 6])

print("\nBroadcasting simples:")
linha = np.array([1.0, 0.0, 2.0, 0.0])
print("b + linha (linha aplicada a cada linha de b):\n", b + linha)
````

## Estatísticas descritivas

Resumimos dados com **tendência central** (média, mediana), **dispersão** (variância, desvio padrão) e **posição** (quartis, percentis). A média é sensível a valores extremos; a mediana costuma ser mais robusta. Com **SciPy**, `scipy.stats.describe` agrega várias estatísticas de uma vez.

````python
amostra = rng.normal(loc=50.0, scale=8.0, size=200)

media = float(np.mean(amostra))
mediana = float(np.median(amostra))
variancia = float(np.var(amostra, ddof=0))
variancia_amostral = float(np.var(amostra, ddof=1))
dp = float(np.std(amostra, ddof=1))

print(f"Média: {media:.3f}")
print(f"Mediana: {mediana:.3f}")
print(f"Variância populacional (ddof=0): {variancia:.3f}")
print(f"Variância amostral (ddof=1): {variancia_amostral:.3f}")
print(f"Desvio padrão amostral (ddof=1): {dp:.3f}")

q25, q75 = np.percentile(amostra, [25, 75])
print(f"Q1 (25%): {q25:.3f} | Q3 (75%): {q75:.3f}")

desc = stats.describe(amostra)
print("\nscipy.stats.describe:")
print(desc)
````

## Correlações

A **correlação linear de Pearson** mede associação linear entre duas variáveis (valores entre -1 e 1). Valores próximos de 1 ou -1 indicam associação linear forte; próximos de 0 indicam fraca linear (mas pode haver relação não linear).

**Atenção:** correlação **não implica causalidade**. Duas variáveis podem subir juntas por causa de um terceiro fator não observado.

Usamos `numpy.corrcoef` para a matriz entre variáveis e `scipy.stats.pearsonr` quando queremos o par (coeficiente, valor-p) em uma amostra.

````python
x = np.array([1, 2, 3, 4, 5, 6], dtype=float)
y = np.array([2.1, 2.9, 4.2, 4.8, 6.0, 7.1])

R = np.corrcoef(x, y)
print("Matriz de correlação 2x2 (Pearson):\n", R)
r_xy = R[0, 1]
print(f"\nCorrelação entre x e y: {r_xy:.4f}")

r_stat, p_valor = stats.pearsonr(x, y)
print(f"pearsonr (SciPy): r = {r_stat:.4f}, p-valor (two-sided) = {p_valor:.4g}")

ruido = rng.standard_normal(len(x))
z = 0.05 * x + ruido
print("\nCorrelação de x com z (sinal fraco + ruído):", float(np.corrcoef(x, z)[0, 1]))
````

## Pequenos exemplos com dados reais

Carregamos um CSV **didático** (`data/vendas_loja.csv`) apenas com **NumPy** (sem Pandas — foco do Módulo 3). O arquivo contém dados **sintéticos** de 12 semanas: gasto em mídias sociais, visitantes e vendas. Consulte a tabela de dados no `README.md` da raiz para a descrição e o aviso de uso didático.

Fluxo típico: carregar → inspecionar dimensões → extrair colunas → estatísticas e correlações.

````python
base = Path.cwd()
csv_path = base / "data" / "vendas_loja.csv"
if not csv_path.exists():
    csv_path = Path("vendas_loja.csv")

raw = np.genfromtxt(csv_path, delimiter=",", names=True, dtype=None, encoding="utf-8")
print("Nomes das colunas:", raw.dtype.names)
print("Número de linhas:", raw.shape[0])

gasto = np.asarray(raw["gasto_midias_sociais_mil_reais"], dtype=float)
visitas = np.asarray(raw["num_visitantes_loja"], dtype=float)
vendas = np.asarray(raw["vendas_mil_reais"], dtype=float)

matriz = np.column_stack([gasto, visitas, vendas])
print("\nPrimeiras 3 linhas (gasto, visitas, vendas):\n", matriz[:3])

print("\nMédias:", matriz.mean(axis=0))
print("Desvios padrão (amostral):", matriz.std(axis=0, ddof=1))

R_all = np.corrcoef(matriz, rowvar=False)
labels = ["gasto_midias", "visitantes", "vendas"]
print("\nMatriz de correlação (Pearson):")
print(np.round(R_all, 3))
print("Variáveis:", labels)
````

## Visualização de dados

O **Matplotlib** é a base clássica de gráficos em Python. Boas práticas: definir **título**, rótulos dos eixos, legenda quando necessário e tamanho de figura legível. Abaixo: **linha** (série temporal das vendas), **dispersão** (visitantes × vendas) e **histograma** (distribuição simulada de uma variável).

````python
semanas = np.arange(1, len(vendas) + 1)

fig, axes = plt.subplots(1, 3, figsize=(12, 3.5))

axes[0].plot(semanas, vendas, marker="o", color="#2563eb")
axes[0].set_title("Vendas por semana")
axes[0].set_xlabel("Semana")
axes[0].set_ylabel("Vendas (mil R$)")
axes[0].grid(True, alpha=0.3)

axes[1].scatter(visitas, vendas, color="#16a34a", edgecolors="black", linewidths=0.4)
axes[1].set_title("Visitantes × vendas")
axes[1].set_xlabel("Número de visitantes")
axes[1].set_ylabel("Vendas (mil R$)")
axes[1].grid(True, alpha=0.3)

axes[2].hist(amostra, bins=15, color="#9333ea", edgecolor="white")
axes[2].set_title("Histograma da amostra simulada")
axes[2].set_xlabel("Valor")
axes[2].set_ylabel("Frequência")

plt.tight_layout()
plt.show()
````
