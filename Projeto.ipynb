## Arrays e operações vetorizadas

O **NumPy** oferece o tipo `ndarray`: arranjos **homogêneos** (mesmo `dtype` em todos os elementos) armazenados de forma eficiente. Isso permite **vetorização**: a mesma operação em muitos valores de uma vez, sem um `for` em Python por posição — em geral mais rápido e mais legível.

**Por que importa em Ciência de Dados?** Colunas numéricas grandes são a regra; somar, normalizar ou comparar vetores inteiros é rotina. O NumPy delega o trabalho pesado a rotinas compiladas.

Conceitos centrais:

- **shape**: tamanho ao longo de cada dimensão (ex.: `(5,)` ou `(3, 4)`).
- **dtype**: tipo dos elementos; evite `dtype=object` sem necessidade.
- **eixos (`axis`)**: em matrizes, `axis=0` agrega sobre linhas (uma estatística por coluna) e `axis=1` sobre colunas.
- **broadcasting**: regras para combinar arrays de formas diferentes (ex.: somar um vetor-linha a cada linha de uma matriz).

![Ideia de broadcasting](img/broadcasting_conceito.png)

Na próxima célula, veja operações vetorizadas e uma máscara booleana antes da seção dedicada a indexação.


````python
# --- Criar arrays e inspecionar metadados ---
lista = [1, 2, 3, 4, 5]
a = np.array(lista, dtype=np.float64)

print("array:", a)
print("shape:", a.shape, "| ndim:", a.ndim, "| dtype:", a.dtype)


# --- Matriz 2D e reshape ---
b = np.arange(12).reshape(3, 4)
print("\nmatriz 3x4:\n", b)


# --- Operações vetorizadas (elemento a elemento) ---
print("\nOperações vetorizadas (elemento a elemento):")
print("a * 10:", a * 10)
print("a ** 2:", a ** 2)
print("sqrt(a):", np.sqrt(a))


# --- Agregações por eixo ---
print("\nAgregações ao longo de eixos:")
print("soma total de b:", b.sum())
print("soma por coluna (eixo 0):", b.sum(axis=0))
print("média por linha (eixo 1):", b.mean(axis=1))


# --- Máscara booleana (filtro sem loop explícito) ---
print("\nMáscaras booleanas (filtrar sem loop):")
print("elementos > 6:", b[b > 6])


# --- Broadcasting ---
print("\nBroadcasting simples:")
linha = np.array([1.0, 0.0, 2.0, 0.0])
print("b + linha (linha aplicada a cada linha de b):\n", b + linha)
````

## Indexação, seleção e filtro em NumPy

**Indexação** escolhe posições; **fatiamento** (`início:fim:passo`) pega intervalos; **máscaras booleanas** filtram por condição; **`np.where`** devolve valores escolhidos por condição, de forma vetorizada.

Em 2D: `M[i, j]` é um elemento; `M[i]` é a linha `i`; `M[:, k]` é a coluna `k`. Índices negativos contam do final (`-1` = último).

![Resumo visual 1D](img/indexacao_numpy_1d.png)

Com **índices avançados** (listas ou arrays de inteiros), `M[[0, 2], [1, 4]]` retorna os elementos nas posições `(0,1)` e `(2,4)` — par a par.


````python
# --- Fatiamento e índices em 1D ---
arr = np.array([10, 20, 30, 40, 50])

print("elemento 0:", arr[0])
print("último:", arr[-1])
print("índices 1 até 3 (exclusivo):", arr[1:4])
print("passo 2:", arr[::2])


# --- Máscara e np.where ---
mascara = arr > 25

print("\nmáscara:", mascara)
print("filtrados:", arr[mascara])

rotulo = np.where(arr >= 30, "alto", "baixo")
print("\nrótulos (np.where):", rotulo)


# --- Matriz: linhas, colunas, bloco ---
M = np.arange(24).reshape(4, 6)

print("\nM:\n", M)
print("\nlinha 2:", M[2])
print("coluna 1:", M[:, 1])
print("bloco [1:3, 2:5]:\n", M[1:3, 2:5])


# --- Indexação avançada (pares linha, coluna) ---
print("\nM[[0, 2], [1, 4]]:", M[[0, 2], [1, 4]])
````

## Filtro, seleção e indexação em NumPy Arrays

Depois de criar arrays, quase sempre precisamos **selecionar subconjuntos** de dados:

- **Indexação posicional**: pegar elemento por índice (`arr[0]`, `arr[-1]`).
- **Fatiamento**: intervalos (`arr[2:8:2]`).
- **Máscara booleana**: filtrar por condição (`arr[arr > 10]`).
- **Indexação avançada**: selecionar várias posições específicas (`arr[[0, 2, 4]]`).

Essas técnicas aparecem o tempo inteiro em análise de dados e pré-processamento.

````python
# --- Bloco 1: vetor base ---
arr = np.array([4, 7, 1, 9, 3, 8, 2, 6, 5, 10])

print("arr:", arr)
print("primeiro elemento arr[0]:", arr[0])
print("último elemento arr[-1]:", arr[-1])


# --- Bloco 2: slicing (fatiamento) ---
print("\narr[2:8]:", arr[2:8])
print("arr[1:9:2] (passo 2):", arr[1:9:2])


# --- Bloco 3: máscaras booleanas (filtro) ---
mascara_maior_5 = arr > 5
print("\nMáscara arr > 5:", mascara_maior_5)
print("Elementos com arr > 5:", arr[mascara_maior_5])

mascara_intervalo = (arr >= 3) & (arr <= 8)
print("Elementos no intervalo [3, 8]:", arr[mascara_intervalo])


# --- Bloco 4: indexação avançada ---
indices = np.array([0, 3, 5, 9])
print("\narr[[0, 3, 5, 9]]:", arr[indices])


# --- Bloco 5: exemplo 2D com seleção por linha/coluna ---
mat = np.arange(1, 17).reshape(4, 4)
print("\nmatriz 4x4:\n", mat)
print("linha 2 (mat[1, :]):", mat[1, :])
print("coluna 3 (mat[:, 2]):", mat[:, 2])
print("submatriz mat[1:3, 1:4]:\n", mat[1:3, 1:4])
````

## Estatísticas descritivas

Resumimos dados com **tendência central** (média, mediana), **dispersão** (variância, desvio padrão) e **posição** (quartis, percentis).

A **média** é sensível a valores extremos; a **mediana** (valor central ordenado) costuma ser mais **robusta** na presença de outliers. A **variância** e o **desvio padrão** medem o quanto os valores se dispersam em torno da média. Em **amostras**, costuma-se usar `ddof=1` na variância/desvio para o estimador não enviesado.

Os **percentis** respondem perguntas do tipo: “abaixo de qual valor ficam 90% dos dados?”. O pacote **SciPy** oferece `scipy.stats.describe` como visão compacta de várias estatísticas.


````python
# --- Amostra simulada ---
amostra = rng.normal(loc=50.0, scale=8.0, size=200)


# --- Tendência e dispersão (NumPy) ---
media = float(np.mean(amostra))
mediana = float(np.median(amostra))
variancia = float(np.var(amostra, ddof=0))
variancia_amostral = float(np.var(amostra, ddof=1))
dp = float(np.std(amostra, ddof=1))

print(f"Média: {media:.3f}")
print(f"Mediana: {mediana:.3f}")
print(f"Variância populacional (ddof=0): {variancia:.3f}")
print(f"Variância amostral (ddof=1): {variancia_amostral:.3f}")
print(f"Desvio padrão amostral (ddof=1): {dp:.3f}")


# --- Quartis ---
q25, q75 = np.percentile(amostra, [25, 75])
print(f"Q1 (25%): {q25:.3f} | Q3 (75%): {q75:.3f}")


# --- SciPy describe ---
desc = stats.describe(amostra)
print("\nscipy.stats.describe:")
print(desc)
````

## Correlações

A **correlação linear de Pearson** mede associação **linear** entre duas variáveis (coeficiente entre -1 e 1). Valores próximos de ±1 indicam alinhamento quase perfeito a uma reta; próximo de 0 indica pouca relação **linear** (ainda pode haver padrão curvo — o gráfico de dispersão ajuda).

![Intuição do coeficiente r](img/correlacao_pearson_intuicao.png)

**Correlação não implica causalidade.** Duas séries podem subir juntas por um terceiro fator não medido.

`numpy.corrcoef` monta a matriz entre variáveis; `scipy.stats.pearsonr` retorna o par (r, p-valor) para testar hipótese sobre correlação nula em amostras — use interpretação cautelosa fora de um curso formal de inferência.


````python
# --- Vetores quase colineares ---
x = np.array([1, 2, 3, 4, 5, 6], dtype=float)
y = np.array([2.1, 2.9, 4.2, 4.8, 6.0, 7.1])

R = np.corrcoef(x, y)
print("Matriz 2x2 (Pearson):\n", R)
print(f"\nr entre x e y: {R[0, 1]:.4f}")


# --- SciPy ---
r_stat, p_valor = stats.pearsonr(x, y)
print(f"pearsonr: r = {r_stat:.4f}, p = {p_valor:.4g}")


# --- Ruído reduz correlação ---
ruido = rng.standard_normal(len(x))
z = 0.05 * x + ruido
print("\ncorrelação x e z:", float(np.corrcoef(x, z)[0, 1]))
````

## Pequenos exemplos com dados reais

Carregamos o CSV didático **`data/vendas_loja.csv`** só com **NumPy** (`np.genfromtxt`). São **12 semanas** de dados **sintéticos**: gasto em mídias, visitantes e vendas. Detalhes e licença de uso estão no `README.md`.

**Pipeline:** (1) ler e conferir nomes das colunas; (2) extrair vetores `float`; (3) calcular médias, desvios e matriz de correlação; (4) na seção seguinte, **visualizar** para validar o que os números sugerem.


````python
# --- Localizar CSV ---
base = Path.cwd()
csv_path = base / "data" / "vendas_loja.csv"

if not csv_path.exists():
    csv_path = Path("vendas_loja.csv")


# --- Leitura com cabeçalho ---
raw = np.genfromtxt(csv_path, delimiter=",", names=True, dtype=None, encoding="utf-8")
print("Colunas:", raw.dtype.names)
print("Linhas:", raw.shape[0])


# --- Colunas numéricas ---
gasto = np.asarray(raw["gasto_midias_sociais_mil_reais"], dtype=float)
visitas = np.asarray(raw["num_visitantes_loja"], dtype=float)
vendas = np.asarray(raw["vendas_mil_reais"], dtype=float)

matriz = np.column_stack([gasto, visitas, vendas])
print("\nPrimeiras 3 linhas:\n", matriz[:3])


# --- Descritivas ---
print("\nMédias:", matriz.mean(axis=0))
print("Desvios (amostral, ddof=1):", matriz.std(axis=0, ddof=1))


# --- Correlações ---
R_all = np.corrcoef(matriz, rowvar=False)
labels = ["gasto_midias", "visitantes", "vendas"]
print("\nCorrelação (Pearson), arredondada:")
print(np.round(R_all, 3))
print("Variáveis:", labels)
````

## Visualização de dados (Matplotlib)

O **Matplotlib** é a base mais comum para gráficos **estáticos** em Python. Boas práticas: título claro, eixos com unidades, grade leve quando ajuda, legenda se houver mais de uma série.

Além de **linha**, **dispersão** e **histograma**, este módulo mostra:

- **Barras** (`bar` / `barh`): comparar categorias ou períodos.
- **Boxplot**: mediana, quartis e outliers — útil para comparar dispersão entre grupos.
- **Pizza (`pie`)**: proporções de um todo. **Uso didático:** funciona com **poucas** fatias; com muitas categorias, prefira barras.
- **Step**: séries que mudam em degraus.
- **`fill_between`**: destacar área sob curva ou entre duas curvas.

Primeiro uma grade **2×3** com linha, dispersão, histograma, barras, barras horizontais e boxplot; depois uma linha com pizza, step e área.


````python
semanas = np.arange(1, len(vendas) + 1)


fig, axes = plt.subplots(2, 3, figsize=(13, 7))
ax = axes.ravel()


# --- Linha ---
ax[0].plot(semanas, vendas, marker="o", color="#2563eb")
ax[0].set_title("Linha — vendas por semana")
ax[0].set_xlabel("Semana")
ax[0].set_ylabel("Vendas (mil R$)")
ax[0].grid(True, alpha=0.3)


# --- Dispersão ---
ax[1].scatter(visitas, vendas, color="#16a34a", edgecolors="black", linewidths=0.4)
ax[1].set_title("Dispersão — visitantes × vendas")
ax[1].set_xlabel("Visitantes")
ax[1].set_ylabel("Vendas (mil R$)")
ax[1].grid(True, alpha=0.3)


# --- Histograma ---
ax[2].hist(amostra, bins=15, color="#9333ea", edgecolor="white")
ax[2].set_title("Histograma — amostra simulada")
ax[2].set_xlabel("Valor")
ax[2].set_ylabel("Frequência")


# --- Barras ---
ax[3].bar(semanas, vendas, color="#ea580c", edgecolor="white")
ax[3].set_title("Barras — vendas por semana")
ax[3].set_xlabel("Semana")
ax[3].set_ylabel("Vendas (mil R$)")


# --- Barras horizontais ---
ax[4].barh(semanas, gasto, color="#0891b2", edgecolor="white")
ax[4].set_title("Barras horiz. — gasto em mídias")
ax[4].set_xlabel("Gasto (mil R$)")
ax[4].set_ylabel("Semana")


# --- Boxplot (visitas em escala reduzida para comparar no mesmo eixo) ---
dados_box = [gasto, visitas / 10.0, vendas]
ax[5].boxplot(dados_box, tick_labels=["Gasto", "Visit./10", "Vendas"])
ax[5].set_title("Boxplot — dispersão (visitas ÷10)")
ax[5].grid(True, axis="y", alpha=0.3)

plt.tight_layout()
plt.show()


# --- Pizza, step, área (segunda figura) ---
fig2, axb = plt.subplots(1, 3, figsize=(13, 3.8))

partes = np.array([float(vendas[:4].sum()), float(vendas[4:8].sum()), float(vendas[8:].sum())])
labels_pie = ["Sem 1–4", "Sem 5–8", "Sem 9–12"]
axb[0].pie(partes, labels=labels_pie, autopct="%1.1f%%", startangle=90)
axb[0].set_title("Pizza — % das vendas por bloco de semanas")

axb[1].step(semanas, vendas, where="mid", color="#7c3aed")
axb[1].set_title("Step — vendas")
axb[1].set_xlabel("Semana")
axb[1].set_ylabel("Vendas")
axb[1].grid(True, alpha=0.3)

axb[2].fill_between(semanas, gasto, alpha=0.35, color="#0d9488", label="Gasto")
axb[2].plot(semanas, gasto, color="#115e59", linewidth=1.5)
axb[2].set_title("Área — gasto (fill_between)")
axb[2].set_xlabel("Semana")
axb[2].set_ylabel("Gasto (mil R$)")
axb[2].legend()

plt.tight_layout()
plt.show()
````
